```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('predictions.csv')
df.head()
```

```python
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    precision_score,
    recall_score,
    f1_score,
    average_precision_score
)

y_true = df['actual']
y_pred = df['predicted']
y_prob = df['fraud_probability']

print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
print("PR-AUC:", average_precision_score(y_true, y_prob))
print("\n Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
print("\n Classification Report:\n", classification_report(y_true, y_pred, digits=4))
```
